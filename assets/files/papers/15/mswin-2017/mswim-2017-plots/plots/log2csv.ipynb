Parse the cooja log output to corret CSV

```python
import os
import sys
```

```python
logDir = "./logs-rpl//"
```

```python
for root, subdirs, files in os.walk(logDir):
    if len(files) > 0:
        toConvertFiles = [f for f in files if "rpl-101-" in f and "log" in f]
        print toConvertFiles   
        print root
        print subdirs
        print files
        for fname in toConvertFiles:
            with open("/".join([root, fname]), "r") as f:
                lines = f.readlines()
                lines = lines[1:-2]
                with open("/".join([root, fname[:-3]])+"csv", "w") as out_f:
                    for l in lines:
                        out_f.write(l)
            os.remove("/".join([root, fname]))
        
        print "\n\n"
    else:
        continue
```
