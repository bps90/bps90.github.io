```python
#process data for log memory AODV

def containsTableInfo(line):
    return "ROUTE_TABLE" in line.split()

def getNode(line):
    idNode = line.split()[1]
    idNode = idNode.replace("(", "")
    idNode = idNode.replace(")", "")
    idNode = idNode.replace(":", "")
    return int(idNode)

def getEntryIDTable(l,):
    return (int(l.split()[6].replace(":", "")) + 1)

def readFile(fileName):
    f = open(fileName, 'r')
    dicNumEntryPerNode = {}
    for l in f.readlines():
        if containsTableInfo(l):
            dicNumEntryPerNode[getNode(l)] = getEntryIDTable(l)
        #print dicNumEntryPerNode
    return dicNumEntryPerNode

        
def sumEntry(dic):
    result = 0
    for key in dic:
        result += dic[key]
    return result
        

        
file_adov_1_flows = "./Logs/Log-memory/AODV-1flows-memory-log-topo-100-random.edgelist"
file_adov_2_flows = "./Logs/Log-memory/AODV-2flows-memory-log-topo-100-random.edgelist"
file_adov_3_flows = "./Logs/Log-memory/AODV-3flows-memory-log-topo-100-random.edgelist"
file_adov_4_flows = "./Logs/Log-memory/AODV-4flows-memory-log-topo-100-random.edgelist"
file_adov_5_flows = "./Logs/Log-memory/AODV-5flows-memory-log-topo-100-random.edgelist"

print sumEntry(readFile(file_adov_1_flows))
print sumEntry(readFile(file_adov_2_flows))
print sumEntry(readFile(file_adov_3_flows))
print sumEntry(readFile(file_adov_4_flows))
print sumEntry(readFile(file_adov_5_flows))

def getNodeWithMaxEntry(dic):
    maxKey = 0
    maxValue = 0
    for k in dic:
        if dic[k] > maxValue:
            maxValue = dic[k]
            maxKey = k
    return maxKey, maxValue


print getNodeWithMaxEntry ( readFile(file_adov_1_flows) )
print getNodeWithMaxEntry ( readFile(file_adov_2_flows) )
print getNodeWithMaxEntry ( readFile(file_adov_3_flows) )
print getNodeWithMaxEntry ( readFile(file_adov_4_flows) )
print getNodeWithMaxEntry ( readFile(file_adov_5_flows) )

```
